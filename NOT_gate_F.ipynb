```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Set seeds for reproducibility
np.random.seed(1)
tf.random.set_seed(1)

# Load the data from the CSV file
file_path = '/content/R1.csv'  # Replace '/content/R2.csv' with your CSV file path
data = pd.read_csv(file_path)

# Assuming 'X1' and 'Y1' are the column names for input and output respectively
X1 = data.iloc[:, 0].values.reshape(-1, 1)  # Reshape for single feature (change accordingly)
Y1 = data.iloc[:, 1].values

# Normalize the input data using StandardScaler
scaler_X = StandardScaler()
scaler_Y = StandardScaler()

X_normalized = scaler_X.fit_transform(X1)
Y_normalized = scaler_Y.fit_transform(Y1.reshape(-1, 1)).flatten()

# Define the neural network model
model = Sequential([
    Dense(64, input_shape=(1,), activation='relu'),  # Adjust input_shape and units as needed
    Dense(32, activation='relu'),
    Dense(1)  # Output layer with 1 neuron for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_normalized, Y_normalized, epochs=1000, batch_size=32, verbose=1)  # Adjust epochs and batch_size as needed

# Make predictions on the entire dataset
predictions_normalized = model.predict(X_normalized)

# Reshape the predictions and true labels to 1D arrays
predictions1 = scaler_Y.inverse_transform(predictions_normalized).flatten()

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(Y1, predictions1)
print(f"Mean Absolute Error (MAE): {mae}")

# Plotting the difference between actual and predicted values
plt.figure(figsize=(10, 5))  # Adjust figure size as needed

# Plot the actual values
plt.scatter(X1, Y1, label='Actual', color='red')  # Change marker color to red

# Plot the predicted values
plt.scatter(X1, predictions1, label='Predicted', color='black')  # Change marker color to black

# Adding labels, title, legend, and grid
plt.xlabel('X1', fontsize=12)  # Adjust font size
plt.ylabel('Y1', fontsize=12)  # Adjust font size
plt.title('Difference between Actual and Predicted Values', fontsize=14)  # Adjust title font size
plt.legend(fontsize='small')  # Adjust legend font size
plt.grid(True, linestyle='--', alpha=0.7)  # Change grid style and transparency

# Display the plot
plt.tight_layout()  # Adjust subplot parameters for a better layout
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Set seeds for reproducibility
np.random.seed(3)
tf.random.set_seed(3)
# Load the data from the CSV file
file_path = '/content/R0.csv'  # Replace '/content/R2.csv' with your CSV file path
data = pd.read_csv(file_path)

# Assuming 'X1' and 'Y1' are the column names for input and output respectively
X = data.iloc[:, 0].values.reshape(-1, 1)  # Reshape for single feature (change accordingly)
Y = data.iloc[:, 1].values

# Normalize the input data using StandardScaler
scaler_X = StandardScaler()
scaler_Y = StandardScaler()

X_normalized = scaler_X.fit_transform(X)
Y_normalized = scaler_Y.fit_transform(Y.reshape(-1, 1)).flatten()

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_normalized, Y_normalized, test_size=0.1, random_state=42)

# Define the neural network model
model = Sequential([
    Dense(64, input_shape=(1,), activation='relu'),
    Dense(32, activation='relu'),
    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, Y_train, epochs=1500, batch_size=16, verbose=1)

# Make predictions on the entire dataset
predictions_normalized = model.predict(X_normalized)

# Reshape the predictions and true labels to 1D arrays
predictions = scaler_Y.inverse_transform(predictions_normalized).flatten()
Y_original = scaler_Y.inverse_transform(Y_normalized.reshape(-1, 1)).flatten()

# Calculate Mean Absolute Error (MAE) on the entire dataset
mae = mean_absolute_error(Y_original, predictions)
print(f"Mean Absolute Error (MAE) on entire dataset: {mae}")

# Plotting the difference between actual and predicted values
plt.figure(figsize=(10, 5))  # Adjust figure size as needed

# Plot the actual values
plt.scatter(scaler_X.inverse_transform(X_normalized), Y_original, label='Actual', color='red')  # Change marker color to red

# Plot the predicted values
plt.scatter(scaler_X.inverse_transform(X_normalized), predictions, label='Predicted', color='black')  # Change marker color to black

# Adding labels, title, legend, and grid
plt.xlabel('r/d', fontsize=12)  # Adjust font size
plt.ylabel('Discharge Coefficient', fontsize=12)  # Adjust font size
plt.title('Difference between Actual and Predicted Values', fontsize=14)  # Adjust title font size
plt.legend(fontsize='small')  # Adjust legend font size
plt.grid(True, linestyle='--', alpha=0.7)  # Change grid style and transparency

# Display the plot
plt.tight_layout()  # Adjust subplot parameters for a better layout
plt.show()
```
