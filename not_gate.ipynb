```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler

# Load the data from the CSV file
file_path = '/content/R1.csv'  # Replace 'your_file.csv' with your CSV file path
data = pd.read_csv(file_path)

# Assuming 'X' and 'Y' are the column names for input and output respectively
X = data.iloc[:, 0].values.reshape(-1, 1) # Reshape for single feature (change accordingly)
Y = data.iloc[:, 1].values


# Normalize the input data using StandardScaler
scaler_X = StandardScaler()
scaler_Y = StandardScaler()

X_normalized = scaler_X.fit_transform(X)
Y_normalized = scaler_Y.fit_transform(Y.reshape(-1, 1)).flatten()

# Define the neural network model
model = Sequential([
    Dense(64, input_shape=(1,), activation='relu'),  # Adjust input_shape and units as needed
    Dense(32, activation='relu'),
    Dense(1)  # Output layer with 1 neuron for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_normalized, Y_normalized, epochs=1500, batch_size=32, verbose=1)  # Adjust epochs and batch_size as needed

# Make predictions
predictions_normalized = model.predict(X_normalized)
predictions = scaler_Y.inverse_transform(predictions_normalized).flatten()

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(Y, predictions)
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
import time  # Import the time module

# Load the data from the CSV file
file_path = '/content/R2.csv'  # Replace '/content/R2.csv' with your CSV file path
data = pd.read_csv(file_path)
start_time = time.time()  # Record the start time

# Assuming 'X1' and 'Y1' are the column names for input and output respectively
X1 = data.iloc[:, 0].values.reshape(-1, 1)  # Reshape for single feature (change accordingly)
Y1 = data.iloc[:, 1].values

# Normalize the input data using StandardScaler
scaler_X = StandardScaler()
scaler_Y = StandardScaler()

X_normalized = scaler_X.fit_transform(X1)
Y_normalized = scaler_Y.fit_transform(Y1.reshape(-1, 1)).flatten()

# Define the neural network model
model = Sequential([
    Dense(64, input_shape=(1,), activation='relu'),  # Adjust input_shape and units as needed
    Dense(32, activation='relu'),
    Dense(1)  # Output layer with 1 neuron for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_normalized, Y_normalized, epochs=1500, batch_size=32, verbose=1)  # Adjust epochs and batch_size as needed
end_time = time.time()  # Record the end time
elapsed_time = end_time - start_time  # Calculate the elapsed time
print(f"Training time: {elapsed_time} seconds")
# Make predictions
predictions_normalized = model.predict(X_normalized)
predictions1 = scaler_Y.inverse_transform(predictions_normalized).flatten()

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(Y1, predictions1)
print(f"Mean Absolute Error (MAE): {mae}")
```
